```python
import pandas as pd, plotly.express as px, seaborn as sns, matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./clean.csv")
```

```python
df.head(3)
```

```python
states_in_nigeria = [
    "Abia", "Adamawa","Akwa Ibom","Anambra",
    "Bauchi","Bayelsa","Benue","Borno",
    "Cross River","Delta",
    "Ebonyi","Edo","Ekiti","Enugu",
    "Gombe","Imo","Jigawa","Kaduna",
    "Kano","Katsina","Kebbi","Kogi","Kwara","Lagos", 
    "Nasarawa","Niger", "Ogun","Ondo", "Osun","Oyo",
    "Plateau","Rivers", "Sokoto","Taraba","Yobe",
    "Zamfara","Federal Capital Territory","Abuja"
]
def is_state_in_nigeria(location):
    for state in states_in_nigeria:
        if state in location:
            return state
    return False
```

```python
df['state'] = df['location'].apply(is_state_in_nigeria)
df['state']
```
